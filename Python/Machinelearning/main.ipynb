## Regression
To find future outcomes or the value at a given point 
### steps for implementing a linear regression
1) Importing req lib (scikit learn)
2) Gathering preprocessed data
3) chosing right model
4) Defining our target(Y-axis) and features(X-axis) 
5) Splitting our data into training and testing ->X_train,X_test,Y_train,Y_test
6) Train our model (X_train,Y_train)
7) Testing our model (X_test,Y_test) -> Validating our models
8) Model evaluation -> accuracy,Precision,F1 scores checks 
9) Visulaization

one target + one feature -> simple linear regression
one target + multi features -> Multiple linear regression
